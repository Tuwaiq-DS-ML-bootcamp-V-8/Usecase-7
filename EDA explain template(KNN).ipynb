### Feature engineering


```python
df.head(2)
```

```python
categorical_features
```

```python
# one hot coding
df = pd.get_dummies(df, columns=categorical_features)
```

```python
df.shape
```

```python
df.head(2)
```

```python
# Calculate the 35th and 75th percentiles of the Current value of yhe player
p35 = df['current_value'].quantile(0.35)
p75 = df['current_value'].quantile(0.75)

# Function to categorize current_values
def categorize_price(current_value):
    if current_value < p35:
        return 'Cheap_Price'
    elif current_value < p75:
        return 'Good_Price'
    else:
        return 'High_Price'

# Apply the function to create a new column
df['player_price_category'] = df['current_value'].apply(categorize_price)

df.drop('current_value', axis=1, inplace=True)

# Verify the distribution of the new categories
print(df['player_price_category'].value_counts())
```

```python
p35, p75
```

```python
df['player_price_category'] 
```

```python
encoder = LabelEncoder()
df['player_price_category_encoded'] = encoder.fit_transform(df['player_price_category'])  
```

### Feature selection

```python
df1 = df
df1.drop(columns=['name',"position_encoded", "winger"], axis=1, inplace=True)
correlation = df1.corr(numeric_only=True)
print(correlation['player_price_category_encoded'].sort_values(ascending=False))
```

```python
# Set the correlation threshold
threshold = 0.2  # You can change this value based on your requirement

# Filter the correlations
# We use `abs()` for absolute value to consider both strong positive and negative correlations
selected_features = correlation[abs(correlation['player_price_category_encoded']) > \
threshold]['player_price_category_encoded'].index
selected_features
```

### Prepare train and test data

```python
# Prepare data
X = df1.drop(['player_price_category_encoded'], axis=1)
y = df1['player_price_category_encoded']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    random_state=42)

# sacle the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X.shape
```

### Buliding the Model¶

```python
model = KNeighborsClassifier()
```

```python
param_grid = {
    'n_neighbors': [2, 3, 4, 5, 6,7,8,9]
}
grid_search = GridSearchCV(estimator=model,
                           param_grid=param_grid,
                           cv=5,
                           scoring='f1_macro', 
                           verbose=1)
```

### Train the Model

```python
# Fit the model on the training data
grid_search.fit(X_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
model = grid_search.best_estimator_
```

### Test the Model

```python
# Predict and evaluate the model
y_pred = model.predict(X_test_scaled)
```

### Evaluating the Model

```python
# our benchmark model
base_model = round(df['player_price_category_encoded'].value_counts()[1]/df.shape[0]*100, 2)
base_model
```

##### 1. Accuracy

```python
# Train score
y_pred_train = model.predict(X_train_scaled)
accuracy = accuracy_score(y_train, y_pred_train)
accuracy
```

```python
# Test score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
print("Train classification report for best knn:")
print(f' {classification_report(y_train, y_pred_train, labels=list(model.classes_))}')
print("Test classification report for best knn: ")
print(f'{classification_report(y_test, y_pred, labels=list(model.classes_))}')
```

##### Confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred,
                                        display_labels=model.classes_,
                                        cmap="Blues",
                                        xticks_rotation='vertical')
plt.title('Confusion Matrix')
plt.show()
```

##### 2. Precision

```python
# Calculate Precision
precision = precision_score(y_test, y_pred, average='macro')
print(f"Precision: {precision:.2f}")
```

##### 3. Recall

```python
# Calculate Recall
recall = recall_score(y_test, y_pred, average='macro')
print(f"Recall: {recall:.2f}")
```

```python
# Calculate F1 Score
f1 = f1_score(y_test, y_pred, average='macro')
print(f"F1 Score: {f1:.2f}")
```
